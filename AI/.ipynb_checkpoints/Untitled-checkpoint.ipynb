```python

# imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


def load_data(path, header):
    marks_df = pd.read_csv(path, header=header)
    return marks_df


if __name__ == "__main__":
    # load the data from the file
    data = pd.read_csv('train.csv')
    data['price_range'] = data['price_range'].map({0:False, 1:False, 2:True, 3:True}).astype(bool)

    x = data.drop('price_range',axis=1)
    y = data['price_range']

    # filter out the applicants that got admitted
    admitted = data.loc[y == True]

    # filter out the applicants that din't get admission
    not_admitted = data.loc[y == False]

    # plots
    plt.scatter(admitted[''], s=10, label='Cheap')
    plt.scatter(not_admitted.iloc[:, 0], not_admitted.iloc[:, 1], s=10, label='Expensive')
    plt.legend()
    plt.show()
```
